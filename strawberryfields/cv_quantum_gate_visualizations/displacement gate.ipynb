## Displacement Gate

Has a specified complex value $\alpha$.

Shifts $\hat{X}$ by value proportional to $Re(\alpha)$ and the $\hat{P}$ by a value $Im(\alpha)$.

```python
import strawberryfields as sf
from strawberryfields.ops import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
prog = sf.Program(1)
eng = sf.Engine('gaussian')

with prog.context as q:
    Dgate(1.25) | q[0]
    
state = eng.run(prog).state
```

```python
fig = plt.figure()

X = np.linspace(-5, 5, 100)
P = np.linspace(-5, 5, 100)
Z = state.wigner(0, X, P)

X, P = np.meshgrid(X, P)

ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, P, Z, cmap=cm.coolwarm, rstride=1, cstride=1, linewidth=0, antialiased=True)

#fig.set_size_inches(4.8, 5)
ax.set_axis_off()

plt.savefig("displacement.png", dpi=1080, transparent=True, bbox_inches='tight', pad_inches=0)
```

As $\alpha$ increases in value, state is displaced more to the right.
