```python
import strawberryfields as sf
from strawberryfields.ops import *

import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
prog = sf.Program(1)
eng = sf.Engine('gaussian')
```

```python
with prog.context as q:
    Vac | q[0]
    
state = eng.run(prog).state
```

```python
X = np.linspace(-5, 5, 100)
P = np.linspace(-5, 5, 100)

Z = state.wigner(0, X, P)
X, P = np.meshgrid(X, P)
```

Plotting 3D Gaussian Wigner function colored with the coolwarm color map.

The surface is made opaque by using antialiased=False

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, P, Z, cmap="magma", lw=0.5, rstride=1, cstride=1, linewidth=0, antialiased=True)

# fig.set_size_inches(4.8, 5)
ax.set_axis_off()

# Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)

# save
plt.savefig("gaussn.png", dpi=1080, transparent=True, bbox_inches='tight', pad_inches=0)

plt.show()
```
