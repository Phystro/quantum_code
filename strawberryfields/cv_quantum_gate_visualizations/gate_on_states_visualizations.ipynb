```python
import strawberryfields as sf
from strawberryfields.ops import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

from mpl_toolkits.mplot3d.axes3d import get_test_data
```

## **Displacement Gate**

### Displacement Gate acting on a Vacuum state

The action of a displacement gate on a vacuum gate with increasing strengths of parameters, $\alpha$

```python
prog = sf.Program(1)   # one qumode circuit
eng = sf.Engine('gaussian')    # initialize engine for circuit

# one qumode vacuum state
with prog.context as q:
    Vac | q[0]
    
# state measurement
state1 = eng.run(prog).state
```

```python
prog = sf.Program(1)   # one qumode circuit
eng = sf.Engine('gaussian')    # initialize engine for circuit

# one qumode vacuum state
with prog.context as q:
    Vac | q[0]
    Dgate(0.75) | q[0]
    
# state measurement
state2 = eng.run(prog).state
```

```python
prog = sf.Program(1)
eng = sf.Engine('gaussian')

with prog.context as q:
    Vac | q[0]
    Dgate(1.50) | q[0]
    
state3 = eng.run(prog).state
```

```python
prog = sf.Program(1)   # one qumode circuit
eng = sf.Engine('gaussian')    # initialize engine for circuit

# one qumode vacuum state
with prog.context as q:
    Vac | q[0]
    Dgate(2.25) | q[0]

# state measurement
state4 = eng.run(prog).state
```

```python
# Initialise the subplot function using number of rows and columns
# fig, axes = plt.subplots(1, 4)

fig = plt.figure(figsize=plt.figaspect(0.24))
#gs = fig.add_gridspec(1, 4, hspace=0, wspace=0)
#(ax1, ax2, ax3, ax4) = gs.subplots(sharex='col', sharey='row')

X = np.linspace(-6, 6, 120)
P = np.linspace(-6, 6, 120)
Z1 = state1.wigner(0, X, P)
Z2 = state2.wigner(0, X, P)
Z3 = state3.wigner(0, X, P)
Z4 = state4.wigner(0, X, P)
X, P = np.meshgrid(X, P)

ax1 = fig.add_subplot(141, projection="3d")
ax2 = fig.add_subplot(142, projection="3d")
ax3 = fig.add_subplot(143, projection="3d")
ax4 = fig.add_subplot(144, projection="3d")
surface1 = ax1.plot_surface(X, P, Z1, cmap='twilight_shifted', lw=0.5, rstride=1, cstride=1, linewidth=0, antialiased=True)
surface2 = ax2.plot_surface(X, P, Z2, cmap='twilight_shifted', lw=0.5, rstride=1, cstride=1, linewidth=0, antialiased=True)
surface3 = ax3.plot_surface(X, P, Z3, cmap='twilight_shifted', lw=0.5, rstride=1, cstride=1, linewidth=0, antialiased=True)
surface4 = ax4.plot_surface(X, P, Z4, cmap='twilight_shifted', lw=0.5, rstride=1, cstride=1, linewidth=0, antialiased=True)
ax1.set_axis_off()
ax2.set_axis_off()
ax3.set_axis_off()
ax4.set_axis_off()

fig.tight_layout(pad=0)

plt.savefig("figures/displacement_sim.png", dpi=300, transparent=True, bbox_inches='tight', pad_inches=0)
```
