## Squeezing Gate

Can be thought of as transformation on position and momentum axes.

```python
import strawberryfields as sf
from strawberryfields.ops import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
prog = sf.Program(1)

with prog.context as q:
    S = Sgate(1)
    S | q[0]
```

```python
eng = sf.Engine('gaussian')
state = eng.run(prog).state
```

```python
fig = plt.figure()

X = np.linspace(-5, 5, 100)
P = np.linspace(-5, 5, 100)

Z = state.wigner(0, X, P)
X, P = np.meshgrid(X, P)

ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, P, Z, cmap=cm.coolwarm, lw=0.5, rstride=1, cstride=1, linewidth=0, antialiased=True)

#fig.set_size_inches(4.8, 5)
ax.set_axis_off()

plt.savefig("squeezed.png", dpi=1080, transparent=True, bbox_inches='tight', pad_inches=0)
```
