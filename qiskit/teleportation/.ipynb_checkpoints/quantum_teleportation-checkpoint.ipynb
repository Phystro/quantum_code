```python
# Do the necessary imports
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, execute, BasicAer, IBMQ
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.extensions import Initialize
# from qiskit_textbook.tools import random_state, array_to_latex
```

```python
def create_bell_pair(qcirc, a, b):
    """Creates a bell pair in the quantum circuit using qubits a and b"""
    qcirc.h(a)
    qcirc.cx(a, b)
```

```python
def alice_gates(qcirc, psi, a):
    """Psi is the qubit being sent to Bob"""
    qcirc.cx(psi, a)
    qcirc.h(psi)
```

```python
def measure_and_send(qcirc, a, b):
    """Measures qubits a and b and sends results to Bob"""
    qcirc.measure(a, 0)
    qcirc.measure(b, 1)
```

```python
def bob_gates(qcirc, qubit, cregz, cregx):
    """The function takes QuantumCircuit, integer(qubit) and
    ClassicalRegisters to decide which gates to apply.
    We use c_if to control our gates with a classical bit instead of
    a qubit."""
    # apply the gates if the registers are in the state '1'
    qcirc.x(qubit).c_if(cregx, 1)
    qcirc.z(qubit).c_if(cregz, 1)
```

```python
# Alice owns q1 and Bob owns q2
# |psi>, psi is the qubit being sent to Bob by Alice, in this case
#     psi is q0
```

```python
# Create Quantum circuit
N = 3    # use 3 qubits
qreg = QuantumRegister(N)
cregx = ClassicalRegister(1, name="cregx")
cregz = ClassicalRegister(1, name="cregz")
qcirc = QuantumCircuit(qreg, cregx, cregz)
```

```python
# initialize q0 in a random state using in initialize gate
# qcirc.x(qreg[0])
# plot_bloch_multivector(psi)

# Step 1
# entangle qubits q1 and q2
create_bell_pair(qcirc, qreg[1], qreg[2])

# Step 2
# apply CNOT to q1, controlled by psi/q0. Then apply H gate to psi/q0
qcirc.barrier()
alice_gates(qcirc, qreg[0], qreg[1])

# Step 3
# apply measurement to both qubits q0 and q1, ,store result in two
# classical bits then send the two bits to Bob
qcirc.barrier()
measure_and_send(qcirc, qreg[0], qreg[1])

# Step 4
# Bob then applies the gates X, Z, or ZX depending on state of the
# classical bits
# if 00 -> do nothing
# if 01 -> apply X-gate
# if 10 -> apply Z-gate
# if 11 -> apply ZX gate
qcirc.barrier()
bob_gates(qcirc, 2, cregz, cregx)

# add a new classical register to see the result
cr_result = ClassicalRegister(1)
qcirc.add_register(cr_result)
qcirc.measure(qreg[2], cr_result)

display( qcirc.draw() )


# simulate
state_backend = BasicAer.get_backend('statevector_simulator')
final_state = execute(qcirc, state_backend).result().get_statevector(qcirc)
plot_bloch_multivector(final_state)

qasm_backend = BasicAer.get_backend('qasm_simulator')
counts = execute(qcirc, qasm_backend, shots=1024).result().get_counts()
plot_histogram(counts)
```
