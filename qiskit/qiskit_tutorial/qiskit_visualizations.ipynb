```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor

from qiskit import Aer, execute
```

```python
bell = QuantumCircuit(2, 2)
bell.h(0)
bell.cx(0, 1)

meas = QuantumCircuit(2, 2)
meas.measure([0, 1], [0, 1])

backend = Aer.get_backend('qasm_simulator')
circ = bell + meas
result = execute(circ, backend, shots=1024).result()
counts = result.get_counts(circ)

print(counts)
```

```python
plot_histogram(counts)
```

```python
second_result = execute(circ, backend, shots=1000).result()
second_counts = second_result.get_counts(circ)

# plot results
leg = ['first execution', 'second execution']
plot_histogram([counts, second_counts], legend=leg)
```

```python
plot_histogram([counts, second_counts], legend=leg, sort='desc', figsize=(15, 12), color=['orange', 'black'], bar_labels=False)
```

### Plot State

```python
from qiskit.visualization import plot_state_city, plot_bloch_multivector
from qiskit.visualization import plot_state_paulivec, plot_state_hinton
from qiskit.visualization import plot_state_qsphere
```

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(bell, backend).result()
psi = result.get_statevector(bell)
```

```python
plot_state_city(psi, title="My City", color=['black', 'orange'], figsize=(19, 10))
```

```python
plot_state_hinton(psi, title="My Hinton", figsize=(19, 10))
```

```python
plot_state_qsphere(psi)
```

```python
plot_state_paulivec(psi, title="Paulivec", color=['purple', 'orange', 'green'])
```

```python
plot_bloch_multivector(psi, title="Bloch Sphere")
```
